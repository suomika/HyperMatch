## 1. HyperMatch vs. Uçar

Here we compare our results from the $\texttt{HgrToGraph}$ programm to the hypergraph matching results of the $\texttt{Karp-Sipser}$ programms based on the Uçar et al. paper.

### 1.1 k-out Hypergraphs

In this section we compare the results of different k-out hypergraphs. The test instances are 2-out, 4-out and 8-out hypergraphs, which differ in the size of nodes (1000, 10000) and in the size of dimensions (3,9).

```python
import numpy as np
import matplotlib.pyplot as plt
import math

kout = ['2', '4', '8']
n1 = 1000
n2 = 10000
```

#### 1.1.1 3-dimensional hypergraphs

```python
# here we have the data for our arrays
ksmd_3dim_n1 = [918/n1, 964/n1, 976/n1]
kss_3dim_n1 = [930/n1, 972/n1, 980/n1]
redumis_3dim_n1 = [962/n1, 984/n1, 994/n1]
wls_3dim_nl = [951/n1, 982/n1, 993/n1]

ksmd_3dim_n2 = [9163/n2, 9624/n2, 9792/n2]
kss_3dim_n2 = [9342/n2, 9696/n2, 9821/n2]
redumis_3dim_n2 = [9570/n2, 9821/n2, 9889/n2]
wls_3dim_n2 = [9537/n2, 9814/n2, 9907/n2]

plt.figure(figsize=(12, 8))

plt.subplot(221)
plt.plot(kout, kss_3dim_n1, 'r',  label="Karp-Sipser scaling")
plt.plot(kout, ksmd_3dim_n1, label="Karp-Sipser minDegree")
plt.plot(kout, redumis_3dim_n1, label="redumis")
plt.plot(kout, wls_3dim_nl, label="weighted local search")
plt.legend(loc="lower right")
plt.title('n = 1000')
plt.xlabel('k = {2, 4, 8}')
plt.ylabel('M_max / n')

plt.subplot(222)
plt.plot(kout, kss_3dim_n2, 'r', label="Karp-Sipser scaling")
plt.plot(kout, ksmd_3dim_n2, label="Karp-Sipser minDegree")
plt.plot(kout, redumis_3dim_n2, label="redumis")
plt.plot(kout, wls_3dim_n2, label="weighted local search")
plt.legend(loc="lower right")
plt.title('n = 10000')
plt.xlabel('k = {2, 4, 8}')
plt.ylabel('M_max / n')

plt.show()
```

#### 1.1.2 9-dimensional hypergraphs

```python
# here we have the data for our arrays
ksmd_9dim_n1 = [591/n1, 630/n1, 672/n1]
kss_9dim_n1 = [526/n1, 572/n1, 617/n1]
redumis_9dim_n1 = [634/n1, 673/n1, 704/n1]
wls_9dim_nl = [621/n1, 659/n1, 702/n1]

ksmd_9dim_n2 = [5885/n2, 6290/n2, 6655/n2]
kss_9dim_n2 = [5246/n2, 5784/n2, 6187/n2]
redumis_9dim_n2 = [6167/n2, 6637/n2, 6905/n2]
wls_9dim_n2 = [6241/n2, 6641/n2, 7015/n2]

plt.figure(figsize=(12, 8))

plt.subplot(221)
plt.plot(kout, kss_9dim_n1, 'r',  label="Karp-Sipser scaling")
plt.plot(kout, ksmd_9dim_n1, label="Karp-Sipser minDegree")
plt.plot(kout, redumis_9dim_n1, label="redumis")
plt.plot(kout, wls_9dim_nl, label="weighted local search")
plt.legend(loc="lower right", prop={'size': 9})
plt.title('n = 1000')
plt.xlabel('k = {2, 4, 8}')
plt.ylabel('M_max / n')

plt.subplot(222)
plt.plot(kout, kss_9dim_n2, 'r', label="Karp-Sipser scaling")
plt.plot(kout, ksmd_9dim_n2, label="Karp-Sipser minDegree")
plt.plot(kout, redumis_9dim_n2, label="redumis")
plt.plot(kout, wls_9dim_n2, label="weighted local search")
plt.legend(loc="lower right", prop={'size': 9})
plt.title('n = 10000')
plt.xlabel('k = {2, 4, 8}')
plt.ylabel('M_max / n')

plt.show()
```

### 1.2 (3-dim) real-life tensors

Here we compare some real-life tensors from the Frostt dataset.

#### 1.2.1 Reduced hypergraphs vs. normal Uçar files

Because the tensors are to big, KaMIS can't handle them. So we had to reduce the number of hyperedges, so that the number of nodes and edges in the resulting graph are not too big. In the first part of this section we compare our results of the reduced hypergraphs with the original tensor files that can be handled by the $\texttt{Karp-Sipser}$ programms of Uçar.

```python
# here we have the data for our arrays
names = ['ksmd', 'kss', 'redumis', 'wls']
uber = [183, 183, 124, 124]
low_uber = min(uber)
high_uber = max(uber)

nips = [2005, 2007, 1422, 1422]
low_nips = min(nips)
high_nips = max(nips)

nell = [5100, 5154, 3037, 3037]
low_nell = min(nell)
high_nell = max(nell)

plt.figure(figsize=(17, 4))

plt.subplot(131)
plt.ylim([math.ceil(low_uber-0.5*(high_uber-low_uber)), math.ceil(high_uber+0.5*(high_uber-low_uber))])
plt.bar(names, uber, color=['blue', 'red', 'orange', 'green'])
plt.title('uber')
plt.ylabel('M_max')

plt.subplot(132)
plt.ylim([math.ceil(low_nips-0.5*(high_nips-low_nips)), math.ceil(high_nips+0.5*(high_nips-low_nips))])
plt.bar(names, nips, color=['blue', 'red', 'orange', 'green'])
plt.title('nips')
plt.ylabel('M_max')

plt.subplot(133)
plt.ylim([math.ceil(low_nell-0.5*(high_nell-low_nell)), math.ceil(high_nell+0.5*(high_nell-low_nell))])
plt.bar(names, nell, color=['blue', 'red', 'orange', 'green'])
plt.title('nell-2')
plt.ylabel('M_max')

plt.show()
```

#### 1.2.2 Reduced hypergraphs vs. reduced hypergraphs

To get a fair comparision, we re-converted the reduced hypergraphs in the Uçar file format. This is possible because in the reduction process, only hyperedges got deleted. So the reduced hypergraphs are still $d$-partit. Logically the results of $\texttt{redumis}$ and $\texttt{weighted_local_search}$ are the same as in section **1.2.1**.

```python
uber_red = [124, 124, 124, 124]
low_uber_red = min(uber_red)
high_uber_red = max(uber_red)

nips_red = [1422, 1422, 1422, 1422]
low_nips_red = min(nips_red)
high_nips_red = max(nips_red)

nell_red = [2994, 3037, 3037, 3037]
low_nell_red = min(nell_red)
high_nell_red = max(nell_red)

plt.figure(figsize=(17, 4))

plt.subplot(131)
plt.ylim(120, 125)
plt.bar(names, uber_red, color=['blue', 'red', 'orange', 'green'])
plt.title('uber')
plt.ylabel('M_max')

plt.subplot(132)
plt.ylim(1418, 1423)
plt.bar(names, nips_red, color=['blue', 'red', 'orange', 'green'])
plt.title('nips')
plt.ylabel('M_max')

plt.subplot(133)
plt.ylim([math.ceil(low_nell_red-0.5*(high_nell_red-low_nell_red)), math.ceil(high_nell_red+0.5*(high_nell_red-low_nell_red))])
plt.bar(names, nell_red, color=['blue', 'red', 'orange', 'green'])
plt.title('nell-2')
plt.ylabel('M_max')

plt.show()
```
